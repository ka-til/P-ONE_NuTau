```python
import numpy as np
import matplotlib.pylab as plt
```

```python
tau_timeDiff = np.loadtxt('20200729_timeDifference_tau.csv')
e_timeDiff = np.loadtxt('20200729_timeDifference_e.csv') 
e_pVal = np.loadtxt('20200729_pval_e.csv')
tau_pVal = np.loadtxt('20200729_pval_tau.csv')
e_wid1_wid2 = np.loadtxt('20200729_wid1_wid2_e.csv')
tau_wid1_wid2 = np.loadtxt('20200729_wid1_wid2_tau.csv')
e_amp1_amp2 = np.loadtxt('20200729_amp1_amp2_e.csv')
tau_amp1_amp2 = np.loadtxt('20200729_amp1_amp2_tau.csv')
```

```python
plt.plot(abs(tau_timeDiff), abs(np.log(tau_pVal)), '.', label='tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_pVal)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('log(p-val(double peak)/p-val(bifurcated gaussian))')
plt.ylim(0, 50)
plt.title('P-Value')
plt.savefig('timediffVSpval1.pdf', dpi = 200)
```

```python
tau_timeDiff.shape, e_timeDiff.shape
```

```python
e_LRR = np.loadtxt('20200729_LRR_e.csv')
tau_LRR = np.loadtxt('20200729_LRR_tau.csv')


plt.plot(abs(tau_timeDiff), abs(np.log(tau_LRR)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_LRR)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('log(LR(double peak)/LR(bifurcated gaussian))')
plt.title('Likelihood Ratio')
plt.savefig('timediffVSLRR1.pdf', dpi = 200)
```

```python
plt.plot(abs(tau_timeDiff), abs(np.log(tau_wid1_wid2)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_wid1_wid2)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('log(wid1/wid2)')
plt.title('Widths of double peak')
plt.savefig('timediffVSwid1.pdf', dpi = 200)
```

```python
plt.plot(abs(tau_timeDiff), abs(np.log(tau_amp1_amp2)), '.', label = 'tau')
plt.plot(abs(e_timeDiff), abs(np.log(e_amp1_amp2)), '.', label = 'e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.ylabel('log(amp1/amp2)')
plt.title('Amps of double peak')
plt.savefig('timediffVSamp1.pdf', dpi = 200)
```

```python
plt.plot(abs(np.log(tau_wid1_wid2)), abs(np.log(tau_LRR)), '.', label = 'tau')
plt.plot(abs(np.log(e_wid1_wid2)), abs(np.log(e_LRR)), '.', label = 'e')
```

```python
bins_tau = np.arange(min(abs(tau_timeDiff)), max(abs(tau_timeDiff))+1, 1)
bins_e = np.arange(min(abs(e_timeDiff)), max(abs(e_timeDiff))+1, 1)
plt.hist(abs(tau_timeDiff), bins = bins_tau, histtype='step', log=True, normed=True, label='tau')
plt.hist(abs(e_timeDiff), bins = bins_e, histtype='step', log=True, normed=True, label='e')
plt.legend()
plt.xlabel('Time Difference (ns)')
plt.savefig('timediffHist1.pdf', dpi=200)
```

```python
bins_tau = np.arange(min(abs(np.log(tau_LRR))), max(abs(np.log(tau_LRR)))+1, 1)
bins_e = np.arange(min(abs(np.log(e_LRR))), max(abs(np.log(e_LRR)))+1, 1)
plt.hist(abs(np.log(tau_LRR)), bins = bins_tau, histtype='step', log=True, normed=False, label='tau')
plt.hist(abs(np.log(e_LRR)), bins = bins_e, histtype='step', log=True, normed=False, label='e')
plt.legend()
plt.xlabel('log(Likelihood Ratio)')
#plt.savefig('likelihood_ratio.pdf', dpi=200)
```

```python
bins_tau = np.arange(min(abs(np.log(tau_amp1_amp2))), max(abs(np.log(tau_amp1_amp2)))+1, 1)
bins_e = np.arange(min(abs(np.log(e_amp1_amp2))), max(abs(np.log(e_amp1_amp2)))+1, 1)
plt.hist(abs(np.log(tau_amp1_amp2)), bins = bins_tau, histtype='step', log=True, normed=True, label='tau')
plt.hist(abs(np.log(e_amp1_amp2)), bins = bins_e, histtype='step', log=True, normed=True, label='e')
plt.legend()
plt.xlabel('Log double peak amplitude ratio')
plt.savefig('amp_ratio1.pdf', dpi=200)
```
