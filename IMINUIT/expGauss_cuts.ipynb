```python
'''
Reading the output file
'''
timeDifferences_t = ([])
timeDifferences_e = ([])

ampRatio_t = ([])
ampRatio_e = ([])

widthRatio_t = ([])
widthRatio_e = ([])

skewnessRatio_t = ([])
skewnessRatio_e = ([])

LRR_t = ([])
LRR_e = ([])

event_number_t = ([])
event_number_e = ([])

largest_2DelLLH_t = ([])
largest_2DelLLH_e = ([])

weight_t = ([])
weight_e = ([])

energy_t = ([])
energy_e = ([])

tLength= ([])
tauInteractionDist = ([])
tau_x = ([])
tau_y = ([])
tauDaughters_x = ([])
tauDaughters_y = ([])
tauDaughters_z = ([])

numHits_t = ([])
numHits_e = ([])

ampAsym_t = ([])
ampAsym_e = ([])

weightPerEventTau = ([])
weightPerEventE = ([])

energyPerEvent_t = ([])
energyPerEvent_e = ([])

tLengthPerEvent = ([])

numHitsPerEvent_t = ([])
numHitsPerEvent_e = ([])

ampAsymPerEvent_t = ([])
ampAsymPerEvent_e = ([])

domPoint_t = ([])
domPoint_e = ([])

wRatio_all = ([])
aRatio_all = ([])
lRatio_all = ([])
kRatio_all = ([])

nFiles = 2000

gcd_file = '/home/users/akatil/P-ONE/GCD_files/PONE_Phase1.i3.gz'
gcd = dataio.I3File(gcd_file)
cframe = gcd.pop_frame()
geometry = cframe["I3Geometry"]
omgeo = geometry.omgeo
print('loaded geometry')

cut = 0
print('CUT:'+str(cut))
event_num = 0
for i in range(0, 2000):
    print('File Num - '+str(i))
    readFile = dataio.I3File('/data/p-one/akatil/step_6_analysis/NuTau_NuE_20Events_expGauss_cuts/step_6_'+str(i)+'_parameters.i3.gz')
    frame_num = 0
    for frame in readFile:
        mctree = frame["I3MCTree"]
        neutrino = frame["NuGPrimary"]
        primary = mctree.primaries
        lepton = dataclasses.I3MCTree.first_child(mctree, primary[0].id)
        params = frame['Parameters_'+str(cut)+'_doublePeak']
        #print(params)
        params_single = frame['Parameters_'+str(cut)+'_biGauss']
        recoPulseMap = frame['I3RecoPulses']
        weight = Weights.weight(frame, nFiles)
        totalHits = numHits.numHits(frame)
        lept_dist = np.sqrt(lepton.pos.x**2+lepton.pos.y**2+lepton.pos.z**2)
        lept_d = dataclasses.I3MCTree.get_daughters(mctree, lepton.id)
        
        muon = 0
        for  ind in range(0, len(lept_d)):
            if lept_d[ind].type == -13 or lept_d[ind].type == 13:
                muon = 1
            if lept_d[ind].type == -16 or lept_d[ind].type == 16:
                td_x = lept_d[ind].pos.x
                td_y = lept_d[ind].pos.y
                td_z = lept_d[ind].pos.z
                
        max2DelLLH = 0
        ampAsym = 0
        point = 0
        dom_point = 0
        for omkey in params.keys():
            oKey = omgeo.get(omkey)
            
            domPos = oKey.position
            x_dom = domPos.x
            y_dom = domPos.y
            z_dom = domPos.z
            
            if lepton.type == 15 or lepton.type == -15:
                first_vertex = np.sqrt((lepton.pos.x - x_dom)**2+(lepton.pos.y - y_dom)**2+(lepton.pos.z - z_dom)**2)
                second_vertex = np.sqrt((td_x - x_dom)**2+(td_y - y_dom)**2+(td_z - z_dom)**2)
            
            recoPulseList = recoPulseMap[omkey]
            recoPulse_timeList = np.array([recoPulse.time for recoPulse in recoPulseList])
            recoPulse_chargeList = np.array([recoPulse.charge for recoPulse in recoPulseList])
            vals = params[omkey]
            vals_single = params_single[omkey]
            
            tDiff = abs(vals[5]-vals[1])
            wRatio = abs(vals[2]-vals[6])
            aRatio = vals[4]/vals[8]
            lRatio = 2*abs(vals_single[0]-vals[0])
            skewRatio = abs(vals[3]-vals[7])
            
            '''
            Introducing Points
            '''    
            ######
            #LLH
            ######
            if lRatio >= 20:
                dom_point=dom_point+np.log10(lRatio)
          
            #print('DOM POINT: ' + str(dom_point))
            
            #wRatio_all = np.append(wRatio_all, width_small/width_large)
            #aRatio_all = np.append(aRatio_all, amp_small/amp_large)
            #lRatio_all = np.append(lRatio_all, lRatio)
            
            ampAsymmetry = abs(vals[4]-vals[8])/(vals[4]+vals[8])
            if lRatio > max2DelLLH:
                max2DelLLH = lRatio
                ampAsym = ampAsymmetry
                point = dom_point
            
            if vals[0] != vals[0] or vals_single[0]!=vals_single[0]:
                print('Found a nan')
                print(vals_single[0], vals[0])
                
            if np.nan in vals or np.nan in vals_single:
                print('Found a nan')
                print(i, frame_num, omkey)
            
            if lepton.type == 15 or lepton.type == -15:
                if np.log10(lepton.length) > 1 and abs(lepton.pos.x) <= 250 and abs(lepton.pos.y) <= 250 and muon==0 and first_vertex < 150 and second_vertex < 150 and abs(first_vertex-second_vertex)<=10:
                    if np.log10(lRatio) > 2:
                        print(lepton.length)
                timeDifferences_t = np.append(timeDifferences_t, tDiff)
                widthRatio_t = np.append(widthRatio_t, wRatio)
                #if dom_point == 3:
                #plots.plot_condition(np.log10(lRatio)>2, vals_single, vals, recoPulse_timeList, recoPulse_chargeList, tDiff, i, frame_num, omkey, lRatio)
                ampRatio_t = np.append(ampRatio_t, aRatio)
                LRR_t = np.append(LRR_t, lRatio)
                skewnessRatio_t = np.append(skewnessRatio_t, skewRatio)
                event_number_t = np.append(event_number_t, event_num)
                numHits_t = np.append(numHits_t, totalHits)
                tLength = np.append(tLength, lepton.length)
                energy_t = np.append(energy_t, neutrino.energy)
                ampAsym_t = np.append(ampAsym_t, ampAsymmetry)
                tauInteractionDist = np.append(tauInteractionDist, lept_dist)
                tau_x = np.append(tau_x, lepton.pos.x)
                tau_y = np.append(tau_y, lepton.pos.y)
                tauDaughters_x = np.append(tauDaughters_x, td_x)
                tauDaughters_y = np.append(tauDaughters_y, td_y)
                tauDaughters_z = np.append(tauDaughters_z, td_z)
                '''
                Calculating the weight of each event
                '''
                weight_t = np.append(weight_t, weight)
            else:
                timeDifferences_e = np.append(timeDifferences_e, tDiff)
                widthRatio_e = np.append(widthRatio_e, wRatio)
                #if dom_point==3:
                plots.plot_condition(lRatio>70, vals_single, vals, recoPulse_timeList, recoPulse_chargeList, tDiff, i, frame_num, omkey, lRatio)
                ampRatio_e = np.append(ampRatio_e, aRatio)
                LRR_e = np.append(LRR_e, lRatio)
                skewnessRatio_e = np.append(skewnessRatio_e, skewRatio)
                event_number_e = np.append(event_number_e, event_num)
                numHits_e = np.append(numHits_e, totalHits)
                energy_e = np.append(energy_e, neutrino.energy)
                ampAsym_e = np.append(ampAsym_e, ampAsymmetry)
                
                '''
                Calculating the weight of each event
                '''
                weight_e = np.append(weight_e, weight)
                
        '''
        Appending 2DelLLH and weight per event
        '''
                
        if lepton.type == 15 or lepton.type == -15:
            largest_2DelLLH_t = np.append(largest_2DelLLH_t, max2DelLLH)
            weightPerEventTau = np.append(weightPerEventTau, weight)
            energyPerEvent_t = np.append(energyPerEvent_t, neutrino.energy)
            tLengthPerEvent = np.append(tLengthPerEvent, lepton.length)
            numHitsPerEvent_t = np.append(numHitsPerEvent_t, totalHits)
            ampAsymPerEvent_t = np.append(ampAsymPerEvent_t, ampAsym)
            domPoint_t = np.append(domPoint_t, dom_point)
        else:
            largest_2DelLLH_e = np.append(largest_2DelLLH_e, max2DelLLH)
            weightPerEventE = np.append(weightPerEventE, weight)
            energyPerEvent_e = np.append(energyPerEvent_e,neutrino.energy)
            numHitsPerEvent_e = np.append(numHitsPerEvent_e, totalHits)
            ampAsymPerEvent_e = np.append(ampAsymPerEvent_e, ampAsym)
            domPoint_e = np.append(domPoint_e, dom_point)
        event_num = event_num+1
        frame_num = frame_num+1
```
